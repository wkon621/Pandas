### **Step 4 Identify missing values**

```python
netflix.isnull().sum() #Director column has the highest num of missing values-1969 cases
```

### **Step 5 Fill the missing values in director column**

```python
netflix['director'].fillna(value='Michale Bay', inplace=True)

netflix['director'].value_counts(dropna=False) #Filled the missing value with Michael Bay becasue I like him 
```

### **Step 6 I want to know the  movies that relaeased in 2019**

```python
movie_2019 = netflix[(netflix.type=='Movie') & (netflix.release_year==2019)] #400 movies were released in 2019
movie_2019
```

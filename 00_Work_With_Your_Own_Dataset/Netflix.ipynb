### **Step 4 Identify missing values**

### **Step 5 Fill the missing values in director column**

```python
netflix['director'].fillna(value='Michale Bay', inplace=True) #Filled the missing value with Michael Bay becasue I like him

netflix['director'].isnull().sum()
```

### **Step 6 I want to know the  movies that relaeased in 2019**

```python
movie_2019 = netflix[(netflix.type=='Movie') & (netflix.release_year==2019)] #400 movies were released in 2019
movie_2019.head(10)
```

### **Step 7 What are the movies that I can watch with my kids regardless of their age?**

```python
movie_2019[(movie_2019.rating=='TV-Y')] #only 9 movies are apprpraite for kids of all age 
```

# **Filtering and Sorting Data** 

###### Find movies.csv from the folder and let's see what we can do to make sense of the dataset ! 

### **Step 1 Import the necessary packages**

```python
import pandas as pd
```

### **Step 2 Assign the dataset to a variable**

```python
movies=pd.read_csv('imdb.csv')
```

### **Step 3 Inspect the first 10 rows** 

```python
movies.head()
```

### **Step 4 Look for movies with longest duration with sort**

```python
#Method 1 Sort only the duration series
movies.duration.sort_values(ascending=False) #476th cell has the longest duration of 242mins
```

```python
#Method 2 Sort by series but output a dataframe
movies.sort_values('duration', ascending=False)
```

### **Step 5 Sort dataframe by multiple columns (star_rating & duration)**

```python
movies.sort_values(['star_rating', 'duration'], ascending=False)
```

### **Step 6 Output a dataframe of movies with duration greater than 200 mins** 

```python
movies[movies.duration>=200]
```

### **Step 6 We want to know the title of movies that are longer than 200 mins** 

```python
#Method 1 
movies[movies.duration>=200]['title']
```

```python
#Method 2
movies[movies.duration>=200].title
```

```python
# Method 3, loc. Loc allows users to select rows and columns by label
movies.loc[movies.duration>=200,'title']
```

### **Step 7  How to add another criteria to filter the data ?**

```python
# look at movies that are longer than 200 mins and are classified as crime by Using "and"

movies[(movies.duration>=200) & (movies.genre=='Crime')]
```

```python
# look at movies that are longer than 200 mins or are classified as drama by Using "or"
movies[(movies.duration>=200) | (movies.genre=='Drama')]
```

### **Step 8 What if you have more than 2 conditions to be fulfilled ?**

```python
#Search for movies that are classified as Drama, Crime, and Adventure

movies[(movies.genre=='Drama')|(movies.genre=='Crime')|(movies.genre=='Adventure')]
```

# **Grouping**

###### Can find the source data drinks.csv from the folder and let's see what we can we do with grouping !

### **Step 1 Import the pandas package**

```python
import pandas as pd
```

### **Step 2 Assign the data to a named dataframe** 

```python
drinks=pd.read_csv('drinks.csv')
```

### **Step 3 Inspect the last 10 occurences**

```python
drinks.tail(10)
```

### Step 4 Which continent drinks more beer on average ?**

```python
drinks.groupby('continent').beer_servings.mean() #Beer consumtion in Europe is the greatest 
```

### **Step 5 What if you only want the beer servings in Asia ?**

```python
drinks[drinks.continent=='Asia'].beer_servings.mean()
```

### **Step 6 Identify the statistics of beer consumption within each continent**

```python
drinks.groupby('continent').beer_servings.describe()
```

### Step 7 What is the mean consumption of the alochol related columns within each continent ? 

```python
drinks.groupby('continent').mean()
```

### **Step 7 What is the macimum consumption of the alochol related columns within each continent?**

```python
drinks.groupby('continent').max()
```

### **Step 8 What if we want to see the combination of count, mean and median of the alochol consumption ?**

```python
drinks.groupby('continent').agg(['count', 'mean', 'median'])
```

# **Fictious Economic Indicator**

### We will create a ficitous data here to explore concatenate  and merge with pandas

### **Step 1 Import necessary package**

```python
import pandas as pd
```

### **Step 2 Create dataframe based on the following data**

```python
ec = {'CPI':[251.1, 252.0, 251.8, 251.3, 251.5], 
        'Int_Rate':[2,3,4,3,2],
       'Inflation_Rate':[1.5,1.5,1.3,1.4, 1.3]}

df=pd.DataFrame(ec)
```

### **Step 3 Inspect the dataframe**

```python
df
```

### **Step 4 Create 3 dataframes based on the following data**

```python
df1 =pd.DataFrame( {'CPI':[251.1, 252.0, 251.8, 251.3, 251.5], 
        'Int_Rate':[2,3,4,3,2],
       'Inflation_Rate':[6,6,7,8,4]},
       index=[2001, 2002, 2003, 2004, 2005])

df2 = pd.DataFrame( {'CPI':[251.3, 252.1, 251.2, 251.8, 251.0],
       'Int_Rate':[2,3,4,3,2],
       'Inflation_Rate':[6,6,7,8,4]},
       index=[2006, 2007, 2008, 2009, 2010])

df3 = pd.DataFrame( {'CPI':[251.1, 252.0, 251.8, 251.3, 251.5],
       'Int_Rate':[2,3,5,5,4],
       'GDP_Capita':[510,1000,480,600,500]},
       index=[2001, 2002, 2003, 2004, 2005])
```

### **Step 5 Lets try concatenate these dataframes**

```python
df4 = pd.concat([df1,df2]) #Concatenation of 2 dataframes
df4
```

```python
df5 = pd.concat([df1,df2,df3], sort=True) #Concatenation of 3 dataframes

df5
```

### **Step 6 What if we want to add rows(series) to the dataframe ?**

```python
a = pd.Series([253, 6, 4.5 ],index= ['CPI', 'Int_Rate', 'Inflation_Rate'])

df4=df.append(a,ignore_index=True)

df4
```

### **Step 7 Merging data along Int_Rate & Inflation_Rate**

```python
df6 = pd.merge(df1,df2, on=['Int_Rate','Inflation_Rate'])
df6
```

### **Step 8 Merge only the data that has the same 'year' on both data1 and data2**

```python
d1 = pd.DataFrame({'Year':[2001,2002,2003,2004],
'Int_rate':[2, 3, 2, 2],
'US_GDP_Thousands':[50, 55, 65, 55]})

d3 = pd.DataFrame({'Year':[2001,2003,2004,2005],
'Unemployment':[7, 8, 9, 6],
'CPI':[250, 252, 250, 253]})

inner=pd.merge(d1,d3, on='Year', how='inner') #year 2001, 2002, and 2004 appeared on both data
inner
```

### **Step 9 Merge all values in data1 and data2 whenever available**

```python
outer=pd.merge(d1,d3, on='Year', how='outer') #Some column show NaN becasue the information can only be found in another data
outer
```
